2.9. создание нового поля вычисленного на основе значений других полей

2.9.1 через выражение на базе имеющихся колонок

```python
# Добавим признак FamilyM - общее число членов семьи на борту
df["FamilyM"] = df["SibSp"] + df["Parch"]
df
```

2.9.2 через DataFrame.apply

```python
df['Pclass'] = df['Pclass'].astype(str)
def function(row):
    if '1' in row['Pclass']:
        result = (row['Fare'] / 3)
        return result
    if '2' in row['Pclass']:
        result = (row['Fare'] / 2)
        return result
    if '3' in row['Pclass']:
        result = row['Fare']
        return result
df['FareMean'] = df.apply(function, axis=1)
df
```

```python
df['Sex'] = df['Sex'].astype(str)
gender_list = list(df['Sex'])
a = pd.Series(gender_list)
def func(value):
    if value == "male":
        return True
    return False
a.apply(func)
```

```python
df['Pclass'] = df['Pclass'].astype(str)
def function(row):
    if '1' in row['Pclass']:
        result = (row['Fare'] / 3)
        return result
    if '2' in row['Pclass']:
        result = (row['Fare'] / 2)
        return result
    if '3' in row['Pclass']:
        result = row['Fare']
        return result
df['FareMean'] = df.apply(function, axis=1)
df.set_index('FareMean', inplace= True )
df
```
